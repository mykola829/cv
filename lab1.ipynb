```python
import matplotlib.pyplot as plt
from matplotlib import image as img

K = [0, 20, 50, 80]

files = [ 
    'high_contrast.jpg', 
    'low_contrast.jpg', 
    'high_detailed.jpg', 
    'low_detailed.jpg'
]
```

```python
def left_shift(fname, k):
    image = img.imread(fname=fname)
    result = image.copy() 
    if(k == 0):
      return result
    for i in range(0, image.shape[0]):
        for j in range(0, image.shape[1]):
            for l in range(0, image.shape[2]):
                result[i][j][l] = min([image[i][j][l] + k, 255])
    return result
```

```python
def plot_images(images, filename) : 
  fig, axs = plt.subplots(1, 4, figsize=(32,18)) 
  for i, ax in enumerate(axs) : 
      ax.imshow(images[i]) 
      ax.set_title(f"{filename}, k = {K[i]}", fontsize=22) 
      ax.axis('off') 
  plt.show()
```

```python
images = []
for file in files:
  for k in K:
    images.append(left_shift(f"sample_data/{file}", k))
  plot_images(images, file.rsplit('.')[0])
  images.clear()
```
